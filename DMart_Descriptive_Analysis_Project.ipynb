# TASK 2:Data Cleaning

```python
#Remove duplicates rows

df.drop_duplicates(inplace=True)
```

```python
#type conversions
df['Bill Number']=df['Bill Number'].astype(str)
```

```python
# check for nulls
df.isnull().sum()
```

```python
df['Order ID'] = df['Order ID'].astype(str)
df['Pin Code'] = df['Pin Code'].astype(str)
```

# TASK 3:Statistics Descriptive

```python
import Arun_theStatistician
Arun_theStatistician.cal_stats('*DISCOUNT PRICE*',df['Discount Price'])
```

```python
Arun_theStatistician.cal_stats('*Maximum Retail Price*',df['MRP'])
```

```python
df['Profit']=df['MRP'] - df['Discount Price']
```

```python
Total_Profit=df['Profit'].sum()
print('Total Profit:',Total_Profit)
```

```python
df['Loss'] = df.apply(lambda row: row['Discount Price'] - row['MRP'] if row['Discount Price'] > row['MRP'] else 0, axis=1)
```

```python
Total_loss=df['Loss'].sum()
print('Total Loss:',Total_loss)
```

# TASK 4:Visualization

```python
#kde=kernal density estimation  ##Distribution plot
sns.displot(df['Discount Price'], kde=True, color='#FF9933')
plt.title('Discount Price Distribution')
plt.show()
```

```python
#distribution plot of MRP
sns.displot(df['MRP'], color='#FF9933')
plt.title('MRP Distribution')
plt.show()
```

```python
# customers are visting store on end-of-month

df['Delivery Date'] = pd.to_datetime(df['Delivery Date'])
sns.displot(df['Delivery Date'].dt.day,color='#138808')
plt.title('Delivery Date Distribution')
```

```python
sns.countplot(data=df,x='Gender',color='#FF9933')
plt.title('Gender Countplot')
plt.show()
# as per graph male visit store more than female
```

```python
sns.countplot(df['Category'],color='#138808')           #this count plot shows that local category is more sold 
plt.title('Category Countplot')
plt.show()
```

```python
# pie chart - State wise Orders

state_counts = df['State'].value_counts()
plt.pie(x=state_counts.values, labels=state_counts.index, autopct='%0.1f%%')
plt.title('order pie chart by state')
```

# task 5: Grouped Analysis

```python
# What is the distribution of customers by age and gender? Are certain age groups more active in purchasing?

sns.displot(df, x='Customer Age', hue='Gender', kind='kde')
plt.title("Customer Age Distribution by Gender")
plt.show()
```

```python
# How does customer subscription status affect total purchase value?

df['Total Order Value'] = pd.to_numeric(df['Total Order Value'], errors='coerce')
subscription_data = df.groupby('Subscription')['Total Order Value'].sum()

plt.figure(figsize=(20, 5))
sns.barplot(x=subscription_data.index, y=subscription_data.values,
            palette=["#FF671F", "#FF9933", "#046A38"], legend=True)
plt.xlabel("Subscription Type")
plt.ylabel("Total Order Value")
plt.title("Total Order Value by Subscription Type")
plt.show()
```

```python
# What are the top-selling products based on total order value?

df.groupby('Product Name')['Total Order Value'].sum()

# Harpic Bathroom Cleaner - Floral is selling at top with revenue of 828936.33 highest among all products
# Parle Nutricrunch Digestive Cookies -Is the second heghest selling product with revenue 821713.48          
```

```python
# What are the top-selling categories based on total order value?

df.groupby('Category')['Total Order Value'].sum()
```

```python
# Which cities and states have the highest number of orders?
state_city = df.groupby('State')['City'].value_counts()
state_city

# Gujrat's Vadodara has highest number of orders
# Maharashtra's Navi Mumbai has highest number of orders
# Andra's Guntur has highest number of orders
# Telanganas's Suryapet has highest number of orders
# at the end Maharshtra state has highest number of orders
```

# Task 6: Correlation Heatmap

```python
df.corr(numeric_only=True)
```

```python
corr_matrix = df[['MRP', 'Discount Price', 'Customer Age', 'Total Order Value']].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, vmax=1, vmin=-1, cmap='viridis', linewidths=0.5, fmt=".2f")

plt.title("Correlation Heatmap")
plt.show()
```
